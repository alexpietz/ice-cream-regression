```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Load the dataset
data = {'Temperatura': [14, 16, 18, 20, 22, 24, 26, 28, 30, 32],
        'Vendas de sorvete': [120, 135, 150, 165, 180, 195, 210, 230, 250, 270]}
df = pd.DataFrame(data)
df
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x='Temperatura', y='Vendas de sorvete', data=df)
plt.title('Temperatura vs Vendas de sorvete')
plt.xlabel('Temperatura (°C)')
plt.ylabel('Vendas (unidades)')
plt.grid()
plt.show()
```

```python
x = df[['Temperatura']]
y = df['Vendas de sorvete']

model = LinearRegression()
model.fit(x, y)
```

```python
y_pred = model.predict(x)
```

```python
rmse = np.sqrt(mean_squared_error(y, y_pred))
r2 = r2_score(y, y_pred)
rmse, r2
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x='Temperatura', y='Vendas de sorvete', s=100, label='Dados reais')
plt.plot(df['Temperatura'], y_pred, color='red', label='Regressão Linear')
plt.title('Regressão Linear: Temperatura vs Vendas de sorvete')
plt.xlabel('Temperatura (°C)')
plt.ylabel('Vendas (unidades)')
plt.legend()
plt.grid()
plt.show()
```
